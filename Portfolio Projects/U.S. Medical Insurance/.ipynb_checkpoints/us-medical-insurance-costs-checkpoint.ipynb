```python
import csv

from pprint import pprint # makes life easier when i can actually see the data
```

I need to make some lists to store the data into

```python
# Indicidual columns
ages = []
sex = []
bmis = []
children = []
isSmoker = []
regions = []
charges = []

# every person in a list please

people = []
```

```python
with open('insurance.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        ages.append(int(row['age']))
        sex.append(row['sex'])
        bmis.append(float(row['bmi']))
        children.append(row['children'])
        isSmoker.append(row['smoker'])
        regions.append(row['region'])
        charges.append(float(row['charges']))
```

```python
for i in range(len(ages)):
    people.append({'age': ages[i], 'sex': sex[i], 'bmi': bmis[i], 'children': children[i], 'smoker': isSmoker[i], 'region': regions[i], 'charges': charges[i] })
```

```python
pprint(people[0]) #just to see columns
```

## Functions/ Work

Make dict with **key = region** and **value = list of people in that area** 

```python
uniqueRegions = set(regions)
print(uniqueRegions)

def create_dict_Region(people,regions):
    dict_byRegion = {}
    for region in regions:
        dict_byRegion[region] = []
    for person in people:
        dict_byRegion[person['region']].append(person)  
    return dict_byRegion
```

```python
people_in_region = create_dict_Region(people,uniqueRegions)
pprint(people_in_region.get('northeast'))
```

## Helpers!!!

A class of helper functions i may need

```python
from scipy import stats
```

```python
class helper:
  # given a list and what you want the averge of, find it
    def dict_find_average(theList,key):
        total = 0
        no_of_items = len(theList) 
        for item in theList:
            total += item[key]
        return total / no_of_items, no_of_items

    def give_array(dictionary,key):
        myArray = []
        for item in dictionary:
            myArray.append(item[key])
        return sorted(myArray)            
        
      
    
helper = helper
```

So now we have a dictionary which split the people up into their respective regions. Now we will investigate Q1 from previous

**1. Average charges for each area**

```python
def create_dict_region_average_charge(regions):
    dict_average_charge = {}
    for region in regions.items():
        average_PeopleTotal = helper.dict_find_average(region[1],'charges')
        IQR = stats.iqr(helper.give_array(region[1],'charges'))
        dict_average_charge[region[0]] = {'average': average_PeopleTotal[0],'IQR': float(IQR), 'IQR-Avg': float(IQR) - average_PeopleTotal[0],  'population': average_PeopleTotal[1]}
    return dict_average_charge
        
    
```

```python
region_average_charge = create_dict_region_average_charge(people_in_region)
pprint(region_average_charge, sort_dicts=False, width=300)
```

The **southeast has the highest average charges**. However, I am aware that the average can be effected by outliers which can both skew the data to trend lower or higher than it should be. Because of this i used the IQR function from the Scipy library to calculate the IQR between the 25th and 75th percentile.

Looking between the Averages and IQRs:

- The IQR of southeast is higher than its average leading me to believe that there are some outlieres that pay way few charges than most people. These anomily/outliers brought the average down. However, looking at the difference these outliers werent too far off.

- Futhermore, we can see that the southeast also has the highest population but odly enough the lowest difference between the IQR and average. So even though one may thing that a higher population may lead to a higher chance of outliers, the oppossite seems to be true here. 

- Looking at the Northwest, Northeast and Southwest we can see that the IQR is less than the average which indicates that all areas had outliers paying substatially  more than the usual which skewed the average higher.

    - The southwest had the largest difference between the IQR and Average which could indicate that it had the highest anomoly in its charges
