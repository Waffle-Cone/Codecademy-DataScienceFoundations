```python
import pandas as pd
```

Import all the files

```python
visits = pd.read_csv('visits.csv',
                     parse_dates=[1])
cart = pd.read_csv('cart.csv',
                   parse_dates=[1])
                   
checkout = pd.read_csv('checkout.csv',
                       parse_dates=[1])
purchase = pd.read_csv('purchase.csv',
                       parse_dates=[1])
```

Step 1: Inspect the DataFrames using `print` and `head`

```python
(visits.head(5))
```

```python
(cart.head(5))
```

```python
(purchase.head(5))
```

```python
(purchase.head(5))
```

Step 2: Left merging visits and cart

```python
visits_cart = pd.merge(visits,cart, how='left')
visits_cart
```

Step 3: How long is `visits_cart`?

```python
len(visits_cart)
```

Step 4: How many timestamps are null for `cart_time`?

```python
not_place_order = visits_cart.cart_time.isnull().sum()
not_place_order
```

Step 5: What percentage only visited?

```python
percent_visit = float(not_place_order)/ float(len(visits_cart))*100

percent_visit
```

Step 6: What percentage placed a t-shirt in their cart but did not checkout?

```python
cart_checkout = pd.merge(cart,checkout,how='left')
cart_checkout
```

```python
no_checkout = cart_checkout.checkout_time.isnull().sum()
no_checkout
```

```python
percent_cart_no_checkout = float(no_checkout)/float(len(cart_checkout))*100
percent_cart_no_checkout
```

Step 7: Merge it all together

```python
all_data = visits_cart.merge(checkout,how='left').merge(purchase,how='left')
all_data.head()
```

Step 8: % of users who got to checkout but did not purchase

```python
got_to_checkout = all_data.checkout_time.count()
purchased = all_data.purchase_time.count()

percent_purchased = float(purchased/got_to_checkout)*100
not_purch = 100 - percent_purchased

not_purch
```

Step 9: check each part of the funnel, let's print all 3 of them again

```python
print("only visit: "+ str(percent_visit))
print("Only add to cart: "+ str(percent_cart_no_checkout))
print("No purchase: "+ str(not_purch))
```

The weakest part of the funnel is getting a person who visited the site to add a tshirt to their cart.

Step 10: adding new column

```python
all_data["total_time"] = all_data.purchase_time - all_data.visit_time
```

Step 11: examine the results

```python
all_data["total_time"]
```

Step 12: average time to purchase

```python
average_time = all_data["total_time"].mean()
```
